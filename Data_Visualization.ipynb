```python
#extracing the different species
setosa_df=iris_df[iris_df.species=='setosa']
vers_df=iris_df[iris_df.species=='versicolor']
virg_df=iris_df[iris_df.species=='virginica']
```

```python
virg_df
```

```python
plt.hist(setosa_df.sepal_width , alpha=0.4, bins=np.arange(2,4.75,0.25))
plt.hist(vers_df.sepal_width ,alpha=0.4, bins=np.arange(2,4.75, 0.25))
#plt.hist(virg_df.sepal_width , bins=np.arange(2,4.75, 0.25))
```

```python
#stacking on top of each other
plt.hist([setosa_df.sepal_width,vers_df.sepal_width,virg_df.sepal_width], bins=np.arange(2, 4.75, 0.25), stacked=True)
```

```python

plt.figure(figsize=(12,8))
plt.hist([setosa_df.sepal_width,vers_df.sepal_width,virg_df.sepal_width], alpha=0.5, bins=np.arange(2,4.75,0.25), stacked=True)
plt.title('Stacked True')
plt.xlabel('yes')
plt.ylabel('no')
plt.legend(['setosa_df','vers_df','virg_df'])
```
